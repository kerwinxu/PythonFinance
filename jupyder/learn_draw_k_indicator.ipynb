```python
_stock_1=sm["sh000001"]
_k=_stock_1.getKData(Query(-200))
#创建两个显示坐标轴的窗口
ax1,ax2 = create_figure(2)

#在第一个坐标轴中绘制K线和EMA
#每个数据都可以绘制图形，都是调用方法plot
#plot参数：
# axes :放在哪个坐标轴
# kdata : k线数据，他会自己计算。
_k.plot(axes=ax1)
ma = EMA(CLOSE(_k))
ma.plot(axes=ax1, legend_on=True)

#在第二个坐标轴中绘制艾尔德力度指标
v = VIGOR(KDATA(_k))
v.plot(axes=ax2, legend_on=True)

```

```python
#创建了3个坐标轴
ax1,ax2, ax3 = create_figure(3)

_k.plot(axes=ax1)
ma.plot(axes=ax1, legend_on=True)
#绘制MACD的方法有2个，
ax_draw_macd(axes=ax2, kdata=_k)
ax_draw_macd2(axes=ax3, ref=ma, kdata=_k)
```
