```python
import tushare as ts
```

```python
_data=ts.get_stock_basics()
```

```python
_data['profit']
```

```python
_data['PEG']=_data['pe']/_data['profit']
```

```python
print("首先看看所有PEG大于0的有多少数量{}".format(len(_data[_data['PEG']>0])))
```

```python
print("所有股票数量{}".format(len(_data)))
```

```python
print("大于0且少于1的数量{}".format(len(_data.loc[(_data['PEG']>0) &(_data['PEG']<1)])))
```

```python
print("我看到那人说成长高，所以最大可以取0.3，这个数量是{}".format(len(_data.loc[(_data['PEG']>0) &(_data['PEG']<0.3)])))
```

```python
print("查看这些股票都是属于什么行业的。可以看到86个行业，也算是很多的了。")
_data.loc[(_data['PEG']>0) &(_data['PEG']<0.3),['industry']].drop_duplicates(['industry'])
```

```python
_data_2=_data.loc[
    (_data['PEG']>0) 
    &(_data['PEG']<0.3) 
    &(_data['name'].str.contains('ST')==False)
    &((_data['area']=="浙江") 
      |(_data['area']=="江苏")
      |(_data['area']=="广东")
      |(_data['area']=="深圳")
      |(_data['area']=="上海")
      |(_data['area']=="福建")
      |(_data['area']=="山东"))
      
]
```
