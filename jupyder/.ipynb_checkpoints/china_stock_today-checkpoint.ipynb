```python
import tushare as ts
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns #要注意的是一旦导入了seaborn，matplotlib的默认作图风格就会被覆盖成seaborn的格式
```

```python
print("查看当天股票相对于昨天的涨跌的")
```

```python
_data=ts.get_day_all()
print("总的股票：{}".format(len(_data)))
_data=_data.loc[(_data['price']!=0)]
print("有效股票：{}".format(len(_data)))
_data
```

```python
_data["p_change_2"]=round(_data["p_change"],1)
```

```python
sns.distplot(_data["p_change_2"])
plt.show()
```

```python
print("平均值：{}".format(_data["p_change"].mean()))
print("上涨的股票个数：{},占比{}%".format(len(_data[_data["p_change"]>0]),len(_data[_data["p_change"]>0])/len(_data)*100))
print("下跌的股票个数：{},占比{}%".format(len(_data[_data["p_change"]<0]),len(_data[_data["p_change"]<0])/len(_data)*100))

print("上涨的平均：{}".format(_data[_data['p_change']>0]['p_change'].mean()))
```
