```python
import tushare as ts
%matplotlib inline
import seaborn as sns #要注意的是一旦导入了seaborn，matplotlib的默认作图风格就会被覆盖成seaborn的格式
import matplotlib.pyplot as plt
from  matplotlib import figure
import pandas as pd
```

# 几个指数的对比

```python
# 取得上证指数
_sh=ts.get_k_data('000001',index=True)
# 取得中证500指数
_zz500=ts.get_k_data('000905',index=True)
# 取得上证50指数
_sz50=ts.get_k_data('000016',index=True)
# 取得中小板指数
_zxb=ts.get_k_data('399005',index=True)

# 转换数据
# 收盘价原先是字符串类型，要转换成浮点类型。
_sh_close=_sh['close'].astype('float')
_zz500_close=_zz500['close'].astype('float')
_sz50_close=_sz50['close'].astype('float')
_zxb_close=_zxb['close'].astype('float')
# 这个日期也要转换，从字符串形式的日期类型转换成真正的日期类型。
_sh_date=pd.to_datetime(_sh['date'])
_zz500_date=pd.to_datetime(_zz500['date'])
_sz50_date=pd.to_datetime(_sz50['date'])
_zxb_date=pd.to_datetime(_zxb['date'])
```

```python
print("查看上证指数和中证500指数的图")
plt.figure(figsize=(18,12))
#为了方便的查看比例，合理用这个来进行缩放。
_ratio_1=_sh_close.max()/_zz500_close.max()
_ratio_2=_sz50_close.max()/_zz500_close.max()
_ratio_3=_zxb_close.max()/_zz500_close.max()

plt.plot(_sh_date,_sh_close/_ratio_1,label='上证指数')
plt.plot(_zz500_date,_zz500_close,label='中证500指数')
plt.plot(_sz50_date,_sz50_close/_ratio_2,label='上证50指数')
plt.plot(_zxb_date,_zxb_close/_ratio_3,label='中小板指')

plt.legend() # 只有加这个才会显示label。
```

```python
print("结果跟教程上的不太一样，我估计原因是，这一年来，他们护住大盘，也就是上证指数的股票，而对于中证500的，则没护住，导致这个结果")
print("可以看到他们护住上证50的力度更大啊。")
```

# 如下是统一成一个函数了

```python
# 我这里建立一个函数吧，方便调用
def plot_zhishu(allow_plot_zhishu):
    # 首先取得相关指数数据的数据啦
    plt.figure(figsize=(18,12))
    _all_zhishu= ts.get_index()
    for _index, _row in _all_zhishu.iterrows():
        if _row['name'] in allow_plot_zhishu:
            # 首先取得指数数据
            _data=ts.get_k_data(_row['code'],index=True)
            # 取得日期
            _date=pd.to_datetime(_data['date'])
            # 取得收盘价
            _close=_data['close'].astype('float')
            #归一化
            _close=_close/(_close.max()-_close.min())
            #绘图
            plt.plot(_date,_close,label=_row['name'])
    plt.legend() # 只有加这个才会显示label。   
            
```

```python
_lst_allow_draw=['上证指数',
                 'Ａ股指数',
                 '上证380',
                '上证180',
                 '上证50',
                 '沪深300',
                '中小板指',
                 '创业板指'
                ]
plot_zhishu(_lst_allow_draw)
```

```python
_lst_allow_draw=['上证指数',
                 'Ａ股指数',
                 '上证380',
                '上证180',
                 '上证50',
                ]
plot_zhishu(_lst_allow_draw)
```
