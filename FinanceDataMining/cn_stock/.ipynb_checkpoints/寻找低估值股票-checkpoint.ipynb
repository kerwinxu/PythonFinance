# 寻找低估值股票1

我这里寻找的，打算从如下几个方面寻找。  
pe：市盈率：表tushare_stock_basics.pe    
pb：市净率：表tushare_stock_basics.pb    
基本每股收益 ：表tushare_stock_basics.esp    
利润同比增长 表：tushare_stock_basics.profit    
流动比率：企业流动资产与流动负债的比率  tushare_stock_debtpaying_data.currentratio  
速动比率：指的是速动资产跟流动负债的比率。速动资产=流动资产 - 存货 - 预付账款 - 待摊费用 tushare_stock_debtpaying_data.quickratio      
roe： 净资产收益率  tushare_stock_report_data.roe  

```python
# 我的做法，我打算先将符合流动比率和速动比率的企业筛选出来
min_currentratio=2
min_quickratio=1
sql_1="""
select code,name, sum(currentratio)/count(currentratio) as currentratio ,
sum(quickratio)/count(quickratio) as quickratio,
count(code) as count_code
from tushare_stock_debtpaying_data
where (year=2018 or year=2017)
group by code
having count_code > 2
and currentratio>{currentratio} 
and quickratio>{quickratio}""".format(currentratio=min_currentratio,quickratio=min_quickratio)
# 市盈率和市净率筛选，这个在总表中。
max_pe=25
max_pb=2
min_profit=0
sql_2="""
select code, pe, pb,profit 
from tushare_stock_basics 
where pe<{pe} and pb<{pb} and profit>{profit}
""".format(pe=max_pe,pb=max_pb,profit=min_profit)
min_roe=4
sql_3="""
select code,sum(roe)/count(roe) as roe,count(code) as count_code
from tushare_stock_report_data
where (year=2018 or year=2017)
group by code
having count_code > 2
and roe>{roe}
""".format(roe=min_roe)
```

```python
# 如上就是3个表的sql。
# 我要把3个表合并成1个。
sql_all="""select s1.code,s1.currentratio,s1.quickratio,s2.pe,s2.pb,s2.profit,s3.roe 
from ({}) as s1 , ({}) as s2 , ({}) as s3 
where s1.code=s2.code and s2.code=s3.code""".format(sql_1,sql_2,sql_3)
```
